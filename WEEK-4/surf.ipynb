```python
import cv2 as cv

# Read the image
image = cv.imread("monkey.jpg")

# Convert to grayscale image
gray_scale = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

# Initialize SIFT object
sift = cv.xfeatures2d.SIFT_create()

# Detect keypoints
keypoints, descriptors = sift.detectAndCompute(gray_scale, None)

# Draw keypoints
sift_image = cv.drawKeypoints(image, keypoints, None)

# Display the image with keypoints
cv.imshow("SIFT Keypoints", sift_image)

# Wait for a key press and then close the window
cv.waitKey(0)
cv.destroyAllWindows()
```
