```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"Mall_Customers.csv")
data.info()
```

```python
x = data.iloc[:,[3,4]].values
""
```

```python
from sklearn.cluster import DBSCAN
model = DBSCAN(eps=7, min_samples=5).fit(x)
labels = model.labels_
```

```python
labels
```

```python
plt.scatter(x[labels == 0, 0], x[labels == 0, 1], s = 100, c = 'blue', label = 'Cluster 0') 
plt.scatter(x[labels == 1, 0], x[labels == 1, 1], s = 100, c = 'green', label = 'Cluster 1')
plt.scatter(x[labels == 2, 0], x[labels == 2, 1], s = 100, c = 'yellow', label = 'Cluster 2') 
plt.scatter(x[labels == 3, 0], x[labels == 3, 1], s = 100, c = 'cyan', label = 'Cluster 3') 
plt.scatter(x[labels == 4, 0], x[labels == 4, 1], s = 100, c = 'magenta', label = 'Cluster 4') 
plt.scatter(x[labels == 5, 0], x[labels == 5, 1], s = 100, c = 'orange', label = 'Cluster 5') 
# plt.scatter(x[labels == -1, 0], x[labels == -1, 1], s = 100, c = 'red', label = 'outliers') 
plt.xlabel("Annual income")
plt.ylabel("spending_score")
plt.legend()
plt.figure(figsize=(10,10))
plt.show()
```
