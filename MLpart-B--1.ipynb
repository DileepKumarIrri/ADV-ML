```python
import pandas as pd
import numpy as np
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
dataset = [['Pizza','Burgur','Ice CReam','Ketch up'],
 ['Pizza','Burgur'],
 ['Pizza','Coke','Garlic Bread'],
 ['Garlic Bread','Coke'],
 ['Garlic Bread','Ketchup'],
 ['Pizza','Coke','Ice cream','Ketchup']]
```

```python
tf = TransactionEncoder()
```

```python
tf_encode = tf.fit_transform(dataset)
tf_encode
```

```python
columns = tf.columns_
print(columns)
```

```python
df = pd.DataFrame(tf_encode, columns = columns)
df
```

```python
famous_items = apriori(df, min_support = 0.2, use_colnames = True)
famous_items
```

```python
association_rules(famous_items, metric='lift')
```
