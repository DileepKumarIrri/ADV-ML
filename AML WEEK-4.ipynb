(ii)Write a program to extract SIFT feature descriptors from a sample image.

```python
import mahotas as mh
import matplotlib.pyplot as plt
```

```python
import cv2
img = cv2.imread(r"D:\MediaGet Downloads\dog image.webp")
# Image1 = mh.colors.rgb2grey(img)
Image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
sift = cv2.SIFT_create()

keypoints, _ = sift.detectAndCompute(Image,None)
```

```python
img=cv2.drawKeypoints(Image ,keypoints ,img ,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
fig = plt.figure(figsize = (10,10))
```

```python
plt.imshow(img)
plt.show()
```
